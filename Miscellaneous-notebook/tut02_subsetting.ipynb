```python
import pandas as pd
```

```python
dataframe = pd.read_csv('train.csv')
print(dataframe.shape)
dataframe.head()
```

```python
dataframe[dataframe.Pclass == 1].tail() # by deualt last 10
# .head() first 10 rows with Pclass = 1
```

```python
dataframe[(dataframe['Pclass']==1) & (dataframe['Sex']=='male')].head()
```

```python
cond1 = dataframe.Pclass > 1
cond2 = dataframe.Sex != 'male'

dataframe[cond1 | cond2]
```

```python
dataframe[(dataframe.Sex == 'male') & (dataframe.Survived == 1) & (dataframe.Age > 28)]
```
