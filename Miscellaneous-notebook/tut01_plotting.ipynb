```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5,5,0.1)
y = x ** 2
plt.plot(x,y)
plt.title('$y=x^2$')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
plt.savefig("figure1.png")
```

```python
# plotting 2
y2 = x ** 3
plt.figure(figsize=(12,5))
plt.plot(x,y,label='$y=x^2$')
plt.plot(x,y2,label='$y = x^3$')
plt.title('powers of x')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()

```
