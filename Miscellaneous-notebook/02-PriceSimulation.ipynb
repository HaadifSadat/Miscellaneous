```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
color_pal = sns.color_palette()
plt.style.use('fivethirtyeight')
```

```python
T = 1000 # number of time steps
P0 = 10 # initial price
mu = 0.001 # drift
```

```python
#last log price
last_p = np.log(P0) 

log_returns = np.zeros(T)
prices = np.zeros(T)

for t in range(T):
    # sample a log return 
    r = 0.01 * np.random.randn() # rand noise

    # compute the new log price
    p = last_p + mu + r

    #store the return and price
    log_returns[t] = r + mu
    prices[t] = np.exp(p)

    # assign last_p
    last_p = p
```

```python
prices = pd.Series(prices)
prices.plot(
    figsize=(20,8),
    title="Simulated Prices",
    color=color_pal[2],
    style="-",
    label="lovda",
)
plt.legend()
plt.show()
```
