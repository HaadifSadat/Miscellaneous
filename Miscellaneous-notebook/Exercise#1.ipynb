### 01 : Generate 1000 sample  form the standard normal

```python
import numpy as np
import matplotlib.pyplot as plt

sample_size = 1000
samples = np.random.randn(sample_size)

print(samples[:10])
```

### 02 : Plot the result as a time series 

```python

sample_size = 1000
samples = np.random.randn(sample_size)

time_index = np.arange(sample_size)


plt.figure(figsize=(10, 5))
plt.plot(time_index, samples)
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Time Series of 1000 Samples from Standard Normal Distribution')
plt.grid(True)
plt.show()
```

### 03 :  Plot the result as a histogram 

```python


sample_size = 1000
samples = np.random.randn(sample_size)

plt.figure(figsize=(10, 5))
plt.hist(samples, bins=30, density=True, alpha=0.7, color='skyblue')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of 1000 Samples from Standard Normal Distribution')
plt.grid(True)
plt.show()
```

### 04 :  Add a trend line to the noise

```python



sample_size = 1000
samples = np.random.randn(sample_size)


trend_slope = 0.01
trend_intercept = 1.0
time_index = np.arange(sample_size)
trend = trend_slope * time_index + trend_intercept


samples_with_trend = samples + trend


plt.figure(figsize=(10, 5))
plt.scatter(time_index, samples_with_trend, s=5, label='Samples with Trend')
plt.plot(time_index, trend, color='red', label='Trend Line')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Scatter Plot with Trend Line')
plt.legend()
plt.grid(True)
plt.show()

```

### call the cumsum() on noise and plot 

```python

sample_size = 1000
samples = np.random.randn(sample_size)


cumulative_sum = np.cumsum(samples)


time_index = np.arange(sample_size)


plt.figure(figsize=(10, 5))
plt.plot(time_index, cumulative_sum)
plt.xlabel('Time')
plt.ylabel('Cumulative Sum')
plt.title('Cumulative Sum of 1000 Samples from Standard Normal Distribution')
plt.grid(True)
plt.show()
```

### Generate and plot 1000 samples from multivariate normal

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the mean vector and covariance matrix for the multivariate normal distribution
mean = np.array([2, 3])   # Mean vector [mean_x, mean_y]
covariance_matrix = np.array([[1, 0.5], [0.5, 1]])  # Covariance matrix

# Generate 1000 samples from the multivariate normal distribution
sample_size = 1000
samples = np.random.multivariate_normal(mean, covariance_matrix, size=sample_size)

x_coords, y_coords = samples[:, 0], samples[:, 1]

plt.figure(figsize=(8, 6))
plt.scatter(x_coords, y_coords, s=10, alpha=0.6, label='Samples')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot of 1000 Samples from Multivariate Normal Distribution')
plt.legend()
plt.grid(True)
plt.show()
```
