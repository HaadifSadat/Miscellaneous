```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
print(df.shape)
df.head()
```

```python
plt.plot(df.Age,df.Fare)
plt.show()
```

```python
plt.scatter(df.Age,df.Fare)
plt.show()
```

```python
df.dropna(inplace=True,subset=['Fare','Age'])
df.shape
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(df.Age.values.reshape(-1,1),df.Fare.values.reshape(-1,1))
x = np.arange(df.Age.min(),df.Age.max(),0.1).reshape(-1,1)
y = model.predict(x)
```

```python
x_Axs = df.Age.values.reshape(-1,1)[:200]
y_Axs = df.Fare.values.reshape(-1,1)[:200]
```

```python
#cohe 
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)

```

```python
y_pred = model.predict(df.Age.values.reshape(-1,1)[-200:])
```

```python
#means squared error
from sklearn.metrics import mean_squared_error
mean_squared_error(y_Axs,y_pred)
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
plt.scatter(x_Axs,y_Axs)
plt.scatter(x_Axs,y_pred)
plt.show()
```
