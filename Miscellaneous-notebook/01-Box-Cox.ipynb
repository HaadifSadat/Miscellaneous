```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import boxcox
color_pal = sns.color_palette()
plt.style.use('fivethirtyeight')
```

```python
# readme CSV
# "Month","International airline passengers: monthly totals in thousands. Jan 49 ? Dec 60"
```

```python
df = pd.read_csv('international-airline-passengers.csv',index_col='Month',parse_dates=True)
df.head()
```

```python
df.plot(
    style="-",
    figsize= (20,8),
    title="Time Series Plot",
    # color=color_pal[4]
    )
plt.show()
```

```python
df['sqrt'] = np.sqrt(df['passengers'])
df['log'] = np.log(df['passengers'])
```

```python
for i in df.columns:
    df[i].plot(
    style="-",
    figsize=(20,8),
    title="Time Series Plot",
    )
    plt.legend()
    plt.show()
```

```python
data , lam = boxcox(df['passengers'])
lam
```

```python
df['BoxCox'] = data
df['BoxCox'].plot(
    figsize = (20,8),
    style="-",
    title="Time Series Split"
)
plt.show()
```

```python
df['passengers'].hist(
    bins=20,
    figsize=(20,8)
    )
plt.show()
```

```python
df['BoxCox'].hist(
    bins=20,
    figsize=(20,8)
    )
plt.show()
```

```python
df['sqrt'].hist(
    bins=20,
    figsize=(20,8)
    )
plt.show()
```
